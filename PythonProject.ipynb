```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

#läser in min fil och som tilldelas variabelnamnet "df". 
df_kpif = pd.read_excel("KPIF-stats12month.xlsx")
```

```python
# Konvertera "Datum"-kolumnen till datumobjekt. Just nu är datumen namngivna "1987M01" för "1987/01"
# Först ersätter vi "M" med ett mellanslag för att underlätta tolkningen och
# och sedan avgränsar datumet till att bara innefatta år och månad
df_kpif["Datum"] = pd.to_datetime(df_kpif["Datum"].str.replace("M", "-")).dt.to_period("M")
        
# Sätter Datumen till index istället för numeriska värden för bättre läsbarhet   
df_kpif.set_index("Datum", inplace=True)
```

```python
value = input("Välj värde: ")

plt.figure(figsize=(15, 6))  # Ställ in storleken på din plot
df_kpif[value].plot(kind="line")  # Får KPIF-värdena som linjediagram
plt.title("KPIF '12 Månaders-utveckling' Utveckling över Tid") #Sätter passande titel
plt.xlabel("Datum") #Har datum på X-axeln
plt.ylabel("KPIF") #Sätter KPIF-värdena på Y-axeln
plt.grid(True) #Lägger till grid så man enklare kan tolka linjen mot Y-axeln
plt.show() #Visar diagrammet
```
